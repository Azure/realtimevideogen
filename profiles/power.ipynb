```python
import pandas as pd

import matplotlib.pyplot as plt
```

# Power
Running Fantasy Talking with different frequency caps.

```python
df_data = pd.read_csv(
    "~/realtimevideogen/profiles/fantasytalking-power.csv",
    header=0,
)
df_data = df_data[~df_data["#run_num"].astype(str).str.startswith("#")]
df_data = df_data.astype({
    "steps": int,
    "width": int,
    "height": int,
    "frequency": float,
    "power": float,
})
df_data
```

```python
filtered = df_data[df_data["steps"] > 0]
plt.plot(
    filtered["frequency"],
    filtered["power"],
    label="Power",
    marker="o")
plt.plot(
    filtered["frequency"],
    filtered["peak_power"],
    label="Peak Power",
    marker="x")
plt.title("Power vs Frequency")
plt.xlabel("Frequency (MHz)")
plt.ylabel("Power (Watts)")
plt.grid(True)
plt.ylim(0, None)
plt.tight_layout(pad=0)
plt.show()
```

```python
plt.plot(
    filtered["frequency"],
    filtered["total"],
    label="Total",
    marker="o")
plt.plot(
    filtered["frequency"],
    filtered["dit_it"] * 4,
    label="DiT",
    marker="x")
plt.plot(
    filtered["frequency"],
    filtered["vae_decode"],
    label="VAE",
    marker="s")
plt.title("Time vs Frequency")
plt.xlabel("Frequency (MHz)")
plt.ylabel("Time (seconds)")
plt.grid(True)
plt.ylim(0, None)
plt.tight_layout(pad=0)
plt.show()
```

```python
# Power vs Time
plt.plot(
    filtered["total"],
    filtered["power"],
    label="Power",
    marker="o")
plt.ylabel("Power (Watts)")
plt.xlabel("Time (seconds)")
plt.title("Power vs Time")
plt.ylim(0, None)
plt.grid(True)
plt.tight_layout(pad=0)
plt.show()
```

```python
# Energy
plt.plot(
    filtered["frequency"],
    filtered["power"] * filtered["total"] / 3600.0,
    label="Energy (Wh)",
    marker="o"
)
plt.title("Energy vs Frequency")
plt.xlabel("Frequency (MHz)")
plt.ylabel("Energy (Wh)")
plt.ylim(0, None)
plt.grid(True)
plt.tight_layout(pad=0)
plt.show()
```
